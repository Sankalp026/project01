## Now , here data is in hourly period of time. so,for better analysis we need to resample our data in monthly

```python
df=df.set_index('Formatted Date')
df.head()
```

```python
df_month_sampling=df[['Apparent Temperature (C)','Humidity']].resample("M").mean()
df_month_sampling.head()
```

```python
plt.figure(figsize=(20,10))
plt.title("Relation of Apparent Temperature and Humidity ")
sns.lineplot(x="Apparent Temperature (C)",y="Humidity",data=df_month_sampling)
```

## plot of apparent Temperature and Humidity with year

```python
plt.figure(figsize=(16,8))
plt.title("plot Apparent Temperature and Humidity with year")
plt.plot(df_month_sampling['Humidity'],label='Humidity',color='Red')
plt.plot(df_month_sampling['Apparent Temperature (C)'],label='Apparent Temperature (C)',color='green')
plt.xlabel("Year",fontsize="14")
plt.legend(loc=5,fontsize=10)
```
